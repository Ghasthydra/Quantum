```
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

def plot_bloch_sphere(state):
    """Plots a Bloch sphere with the given state vector."""

    # Extract Bloch vector components
    bloch_vector = [
        2 * np.real(state[0] * np.conj(state[1])),  # X component
        2 * np.imag(state[0] * np.conj(state[1])),  # Y component
        np.abs(state[0])**2 - np.abs(state[1])**2   # Z component
    ]

    # Create the Bloch sphere
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Draw the sphere
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x = np.cos(u) * np.sin(v)
    y = np.sin(u) * np.sin(v)
    z = np.cos(v)
    ax.plot_wireframe(x, y, z, color="k", alpha=0.1)

    # Draw the Bloch vector
    ax.quiver(0, 0, 0, bloch_vector[0], bloch_vector[1], bloch_vector[2],
              color="r", arrow_length_ratio=0.1)

    # Set axis limits and labels
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    # Show the plot
    plt.show()

# Example state vector (|0⟩ state)
state = np.array([1, 0])
plot_bloch_sphere(state)
```
